# 4. Get the number of clusters

## 4.1 Elbow method

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
X = data_df.copy()
#X.drop('id', axis=1, inplace=True)

model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,20), metric='distortion', timings=True)
visualizer.fit(X)
visualizer.show()
```

```python
# Number of clusters: k=6
```

# 5. Generate clusters

## 5.1 BayesianGaussianMixture

K-means method is not very good, and the clusters might be not circular nor equaly densed. We'll use a probabilistic model (BayesianGaussianMixture)

```python
CLUSTERS = 7

model = BayesianGaussianMixture(
    n_components=CLUSTERS,
    max_iter=300,
    n_init=10,
    covariance_type='full',
    random_state=101,
    verbose_interval=100,
)

predicted_clusters = model.fit_predict(X)
predicted_proba = model.predict_proba(X)

X['predicted'] = predicted_clusters
```

```python
X.head()
```

## 5.2 BayesianGMMClassifier

```python
# Get the max probability
max_proba = predicted_proba.max(axis=1)
X['max_proba'] = max_proba
```

```python
# Set threshold ti 75%
data = X[X['max_proba']>0.75]
```

```python
data
```

```python
data_gmm = data.drop(columns=['max_proba'], axis=1)
X_gmm = data_gmm.drop(columns='predicted',axis=1)
y_gmm = data_gmm['predicted']
```

```python
model_gmm = BayesianGMMClassifier(
    n_components=CLUSTERS, 
    tol=1e-3, 
    covariance_type='full', 
    max_iter=400, 
    n_init=4, 
    init_params='kmeans', 
    random_state=101)

model_gmm.fit(X_gmm, y_gmm)
clusters_pred =  model_gmm.predict(data_df)
```

# 6. Make output file

```python
# Load the dataset
submission = pd.read_csv('sample_submission.csv', sep=',', encoding='UTF-8', index_col='Id')
submission['Predicted'] = clusters_pred
submission.to_csv('submission.csv')
```
